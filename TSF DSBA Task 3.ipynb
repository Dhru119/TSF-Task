# Prediction using Decision Tree Algorithm

Creating the Decision Tree classifier for the Iris dataset and visualizing it graphically so that if we feed any new data to this classifier, it would be able to predict the right class accordingly.

```python
#Importing Required Libraries and Datasets
import pandas as pd
from sklearn import datasets
```

```python
#Reading and Loading the IRIS Dataset
iris = datasets.load_iris()
i_data = pd.DataFrame(iris.data, columns = iris.feature_names)
print(i_data.shape)
i_data.head()
```

```python
y=iris.target
print(y)
```

Now, we will define the Decision Tree Algorithm and apply it to our Iris dataset.

```python
#Importing the Decision Tree Classifier and fitting the IRIS Dataset
from sklearn.tree import DecisionTreeClassifier
clf=DecisionTreeClassifier()
clf.fit(i_data,y)
```

So, it's time now to visualise the Decision Tree graphically to analyse the given data.

```python
#Importing Necessary Libraries for observation
from sklearn.tree import export_graphviz
from six import StringIO  
from IPython.display import Image  
import pydotplus
```

```python
#Visualising the Decision Tree for better Understanding 
dot_d = StringIO()
export_graphviz(clf, out_file=dot_d, feature_names=iris.feature_names,  filled=True, special_characters=True)
graph = pydotplus.graph_from_dot_data(dot_d.getvalue())  
Image(graph.create_png())
```

So, if any new data is now fed to this Decision Tree Classifier, the model will be able to predict the right class in which it belongs.
